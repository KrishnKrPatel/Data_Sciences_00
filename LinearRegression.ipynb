```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy import stats
```

```python
x=np.random.normal(3.0,1.0,1000)
y=100-(x+np.random.normal(0.0,0.1,1000))*3
```

```python
plt.scatter(x,y,color='red')
plt.show();
```

```python
slope,intercept,r_value,p_value,std_err=stats.linregress(x,y)
```

```python
r_value
```

```python
r_value**2
```

```python
def linear_Regression(data):
    return slope*data+intercept

fitline=linear_Regression(x)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,fitline,color='blue')
plt.show()
```

# Polynomial Regression

```python
np.random.seed(2)
```

```python
x=np.random.normal(3.0,1.0,1000)
y=np.random.normal(50.0,10.0,1000)/x
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from scipy import stats
```

```python
x=np.array(x)
y=np.array(y)
```

```python
p4=np.poly1d(np.polyfit(x,y,4))
```

```python
xp=np.linspace(0,7,100)
```

```python
plt.scatter(x,y,color='red')
plt.plot(xp,p4(xp),color='blue')
plt.show()
```
